```python
X_train=X_train.drop(['Name','Ticket'],axis=1)
```

```python
X_train=X_train.drop(['Embarked','Cabin'],axis=1)
```

```python
X_train
X_train.Sex[X_train.Sex == 'male'] = 1
X_train.Sex[X_train.Sex == 'female'] = 2
X_train=X_train.drop(['PassengerId'],axis=1)
```

```python
X_train=X_train.dropna()
```

```python
y_train=X_train[['Survived']]
```

```python
X_train= X_train.drop(['Survived','Age','Fare'],axis=1)
```

```python
y_train
```

```python
regr = LogisticRegression(random_state=0)
```

```python
X_train
```

```python

clf = regr.fit(X_train, y_train)
```

```python
clf.score(X_train,y_train)
```
